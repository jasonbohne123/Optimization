### Gradient Descent 

```python
import numpy as np
# ^^^ pyforest auto-imports - don't write above this line
import numpy as np
import autograd
import matplotlib.pyplot as plt
```

Take our function on the domain below, which is continuous and differentiable

```python
x=np.linspace(-4,4,100)
def f(x):
    return x**4-x**2 +10*x
plt.plot(x,f(x),label='Original')
x_deriv=np.array([autograd.grad(f)(i) for i in x])
plt.plot(x,x_deriv,color='orange',label='AutoGrad Derivative')
plt.legend()
```

Say we wish to find the minimum of this function. We can implement Gradient Descent

```python
def gradient_descent(x,function,learning_rate,max_iter=500,tol=0.001):
    x=float(x)
    x_vals=np.array([])
    gradient=autograd.grad(function) # auto-differentiate function
    
    for i in range (0,max_iter):
        correction=learning_rate*gradient (x)
        x_vals=np.append(x_vals,x)
        if abs(correction) <tol:
            print("Min Tol Reached at ",i)
            return x,x_vals
        x=x-correction
    print("Max Iter reached")
    return x,x_vals
```

```python
x_min,x_vals=gradient_descent(3.0,f,0.025,max_iter=5000)
```

```python
plt.plot(x,f(x))
plt.plot(x_vals,f(x_vals),color='red')
plt.title("Gradient Descent Iterations")
```

```python
x_min,x_vals=gradient_descent(-4.0,f,0.025,max_iter=5000)
```

```python
plt.plot(x,f(x))
plt.plot(x_vals,f(x_vals),color='red')
plt.title("Gradient Descent Iterations")
```

### Multiple Local Optimums

```python
x=np.linspace(-4,4,100)
def f(x):
    return x**4-20*x**2 +10*x

plt.plot(x,f(x))
```

```python
x_min,x_vals=gradient_descent(0,f,0.025,max_iter=5000,tol=0.1)
```

```python
plt.plot(x,f(x))
plt.plot(x_vals,f(x_vals),color='red')
plt.title("Gradient Descent Iterations")
```

```python
x_min,x_vals=gradient_descent(0.5,f,0.025,max_iter=5000,tol=0.1)
```

```python
plt.plot(x,f(x))
plt.plot(x_vals,f(x_vals),color='red')
plt.title("Gradient Descent Iterations")
```
